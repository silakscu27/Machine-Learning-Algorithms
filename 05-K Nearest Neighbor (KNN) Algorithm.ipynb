```python
from sklearn.model_selection import train_test_split
```

```python
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=0)
from sklearn.neighbors import KNeighborsClassifier
snf=KNeighborsClassifier(n_neighbors=3)
snf.fit(X_train,y_train)
```

```python
snf.predict(X_test)
```

```python
snf.score(X_test,y_test)
```

**PRACTICE**

**Importing data**

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

**Exploring the data**

```python
print(cancer['DESCR'])
```

**Preprocessing the data**

```python
X_train, X_test, y_train, y_test=train_test_split(cancer.data, cancer.target,
                                                  stratify=cancer.target, random_state=0)
```

```python
train_accuracy=[]
test_accuracy=[]
```

```python
number_of_nghb=range(1,11)
```

```python
import matplotlib.pyplot as plt

for n_nghb in number_of_nghb:
    snf=KNeighborsClassifier(n_neighbors=n_nghb)
    snf.fit(X_train,y_train)
    train_accuracy.append(snf.score(X_train,y_train))
    test_accuracy.append(snf.score(X_test, y_test))

plt.plot(number_of_nghb, train_accuracy, label="train accuracy")
plt.plot(number_of_nghb, test_accuracy, label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n-neighbors")
plt.legend()
```
